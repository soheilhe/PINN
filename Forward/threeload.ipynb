```python
import deepxde as dde
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import time
```

```python

L = 5
pl1 = 1
pl2 = 2.5
pl3 = 4


p = 5*10**5
E = 24855578000
I = 0.00045


#convert Newton to MegaNewton
p = p*10**-6
E = E*10**-6

scale = 1

EI = E*I


# Define the domain

geom = dde.geometry.Interval(0, L)


def dy(x, y):
    return dde.grad.jacobian(y, x)

def ddy(x, y):
    return dde.grad.hessian(y, x)

def dddy(x, y):
    return tf.gradients(ddy(x, y), x)[0]

def pde(x, y):
    x_left = tf.greater(x, pl1)
    x_right = tf.less(x, pl2)
    y_left = tf.greater(x, pl2)
    y_right = tf.less(x, pl3)

    left = tf.less(x, pl1)
    right = tf.greater(x , pl3)
    mid1 = x_left & x_right
    mid2 = y_left & y_right

    y1 = tf.where(left, y[:,0:1], y[:,1:4])
    y2 = tf.where(mid1, y[:,1:2], tf.concat([y[:, 0:1],y[:, 2:4]], axis=1))
    y3 = tf.where(mid2, y[:,2:3], tf.concat([y[:, 0:2],y[:, 3:4]], axis=1))
    y4 = tf.where(right, y[:,3:4],y[:,0:3])

    dy1_x = dde.grad.jacobian(y1, x)
    dy1_xx = dde.grad.jacobian(dy1_x, x)
    dy1_xxx = dde.grad.jacobian(dy1_xx, x)
    dy1_xxxx = dde.grad.jacobian(dy1_xxx, x)
    dy2_x = dde.grad.jacobian(y2, x)
    dy2_xx = dde.grad.jacobian(dy2_x, x)
    dy2_xxx = dde.grad.jacobian(dy2_xx, x)
    dy2_xxxx = dde.grad.jacobian(dy2_xxx, x)
    dy3_x = dde.grad.jacobian(y3, x)
    dy3_xx = dde.grad.jacobian(dy3_x, x)
    dy3_xxx = dde.grad.jacobian(dy3_xx, x)
    dy3_xxxx = dde.grad.jacobian(dy3_xxx, x)
    dy4_x = dde.grad.jacobian(y4, x)
    dy4_xx = dde.grad.jacobian(dy4_x, x)
    dy4_xxx = dde.grad.jacobian(dy4_xx, x)
    dy4_xxxx = dde.grad.jacobian(dy4_xxx, x)

    return tf.where(left, EI*dy1_xxxx, tf.where(mid1, EI*dy2_xxxx, tf.where(mid2, EI*dy3_xxxx, EI*dy4_xxxx)))


# Boundary conditions

def boundary_l(x, on_boundary):
    return on_boundary and np.isclose(x[0], 0)
def boundary_r(x, on_boundary):
    return on_boundary and np.isclose(x[0], L)

def on_interf_1(x, _):
    return np.isclose(x[0], pl1)
def on_interf_2(x, _):
    return np.isclose(x[0], pl2)
def on_interf_3(x, _):
    return np.isclose(x[0], pl3)

def data_1(x, _):
    return np.isclose(x[0], L/2)
def data_2(x, _):
    return np.isclose(x[0], L/8)
def data_3(x, _):
    return np.isclose(x[0], 7*L/8)
def data_4(x, _):
    return np.isclose(x[0], L/4)
def data_5(x, _):
    return np.isclose(x[0], 3*L/4)
def data_6(x, _):
    return np.isclose(x[0], 3*L/8)
def data_7(x, _):
    return np.isclose(x[0], 5*L/8)

xxxx=np.array(([pl1-0.01],[pl1],[pl1+0.01],[pl2-0.01],[pl2],[pl2+0.01],[pl3-0.01],[pl3],[pl3+0.01],[L/2],[L/4],[3*L/4],[L/8],[7*L/8],[3*L/8],[5*L/8],[0],[L],[2*L/20],[18*L/20],[3*L/20],[17*L/20],[2*L/10],[8*L/10],[6*L/10],[4*L/10]))

def continuity_deflection(x, y,_):
    y1 = y[:,0:1]
    y2 = y[:,1:2]
    y3 = y[:,2:3]
    y4 = y[:,3:4]
    return [y1,y2,y3,y4]

def continuity_slope(x, y,_):
    y1 = y[:,0:1]
    y2 = y[:,1:2]
    y3 = y[:,2:3]
    y4 = y[:,3:4]
    dy1_x = dde.grad.jacobian(y1, x)
    dy2_x = dde.grad.jacobian(y2, x)
    dy3_x = dde.grad.jacobian(y3, x)
    dy4_x = dde.grad.jacobian(y4, x)
    return [dy1_x,dy2_x,dy3_x,dy4_x]

def continuity_moment(x, y, X):
    y1 = y[:,0:1]
    y2 = y[:,1:2]
    y3 = y[:,2:3]
    y4 = y[:,3:4]
    dy1_x = dde.grad.jacobian(y1, x)
    dy1_xx = dde.grad.jacobian(dy1_x, x)
    dy2_x = dde.grad.jacobian(y2, x)
    dy2_xx = dde.grad.jacobian(dy2_x, x)
    dy3_x = dde.grad.jacobian(y3, x)
    dy3_xx = dde.grad.jacobian(dy3_x, x)
    dy4_x = dde.grad.jacobian(y4, x)
    dy4_xx = dde.grad.jacobian(dy4_x, x)
    return [dy1_xx,dy2_xx,dy3_xx,dy4_xx]

def continuity_shear(x, y, X):
    y1 = y[:,0:1]
    y2 = y[:,1:2]
    y3 = y[:,2:3]
    y4 = y[:,3:4]
    dy1_x = dde.grad.jacobian(y1, x)
    dy1_xx = dde.grad.jacobian(dy1_x, x)
    dy1_xxx = dde.grad.jacobian(dy1_xx, x)
    dy2_x = dde.grad.jacobian(y2, x)
    dy2_xx = dde.grad.jacobian(dy2_x, x)
    dy2_xxx = dde.grad.jacobian(dy2_xx, x)
    dy3_x = dde.grad.jacobian(y3, x)
    dy3_xx = dde.grad.jacobian(dy3_x, x)
    dy3_xxx = dde.grad.jacobian(dy3_xx, x)
    dy4_x = dde.grad.jacobian(y4, x)
    dy4_xx = dde.grad.jacobian(dy4_x, x)
    dy4_xxx = dde.grad.jacobian(dy4_xx, x)
    return [dy1_xxx,dy2_xxx,dy3_xxx,dy4_xxx]


#Two end supports

bc1 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[0], boundary_l)
bc2 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_slope(x, y, _)[0], boundary_l)
bc3 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[3] , boundary_r)
bc4 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_slope(x, y, _)[3] , boundary_r)

bc5 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[0] - continuity_deflection(x, y, _)[1] , on_interf_1)
bc6 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_slope(x, y, _)[0] - continuity_slope(x, y, _)[1] , on_interf_1)
bc7 = dde.icbc.OperatorBC(geom, lambda x, y, _: EI*continuity_moment(x, y, _)[0] - EI*continuity_moment(x, y, _)[1] , on_interf_1)
bc8 = dde.icbc.OperatorBC(geom, lambda x, y, _: EI*continuity_shear(x, y, _)[0] -  EI*continuity_shear(x, y, _)[1] - p , on_interf_1)

bc9 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[1] - continuity_deflection(x, y, _)[2] , on_interf_2)
bc10 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_slope(x, y, _)[1] - continuity_slope(x, y, _)[2] , on_interf_2)
bc11 = dde.icbc.OperatorBC(geom, lambda x, y, _: EI*continuity_moment(x, y, _)[1] - EI*continuity_moment(x, y, _)[2] , on_interf_2)
bc12 = dde.icbc.OperatorBC(geom, lambda x, y, _: EI*continuity_shear(x, y, _)[1] -  EI*continuity_shear(x, y, _)[2] - p , on_interf_2)

bc13 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[2] - continuity_deflection(x, y, _)[3] , on_interf_3)
bc14 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_slope(x, y, _)[2] - continuity_slope(x, y, _)[3] , on_interf_3)
bc15 = dde.icbc.OperatorBC(geom, lambda x, y, _: EI*continuity_moment(x, y, _)[2] - EI*continuity_moment(x, y, _)[3] , on_interf_3)
bc16 = dde.icbc.OperatorBC(geom, lambda x, y, _: EI*continuity_shear(x, y, _)[2] -  EI*continuity_shear(x, y, _)[3] - p , on_interf_3)

ic1 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[1] + 0.049592 , data_1) #L/2 ,2.5
ic2 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[0] + 0.00971323 , data_2) #L/8 ,0.625
ic3 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[3] + 0.00971323 , data_3) #7*L/8 ,4.375
ic4 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[1] + 0.0280556 , data_4) #L/4 ,1.25
ic5 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[2] + 0.0280556 , data_5) #3*L/4 ,3.75
ic6 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[1] + 0.0432985 , data_6) #3*L/8 ,1.875
ic7 = dde.icbc.OperatorBC(geom, lambda x, y, _: continuity_deflection(x, y, _)[2] + 0.0432985 , data_7) #5*L/8 ,3.125

def func(x):
    a = 1
    b = 2.5
    c = 4
    if x<a:
        return (-0.0318507*x**2 + 0.0111757*x**3)
    elif a<=x<b:
        return (0.00745045 - 0.0223513*x - 0.00949932*x**2 + 0.00372522*x**3)
    elif b<=x<c:
        return (0.123864 - 0.162047*x + 0.046379*x**2 - 0.00372522*x**3)
    elif x>=c:
        return (0.600692 - 0.519669*x + 0.135784*x**2 - 0.0111757*x**3)

boundary_condition_list = [bc16,ic1,ic2,ic3,ic4,ic5,ic6,ic7]

# Create empty lists to store results
results = []
losses = []

layer_size = [1] + [20] * 3 + [4]
activation = "tanh"
initializer = "Glorot uniform"
net = dde.nn.FNN(layer_size, activation, initializer)

boundary_conditions = [bc1, bc2, bc3, bc4, bc5, bc6, bc7, bc8, bc9, bc10, bc11, bc12, bc13, bc14, bc15]

for bc in boundary_condition_list:
    boundary_conditions.append(bc)

    start_time = time.time()
    data = dde.data.PDE(
        geom,
        pde,
        boundary_conditions,
        num_domain=0,
        num_boundary=0,
        #solution=func,
        anchors=xxxx,
    )

    model = dde.Model(data, net)
    model.compile("adam", lr=0.0005)

    # Initialize variables
    iterations_list = [1000, 2000, 2000, 5000, 10000, 20000]

    for iterations in iterations_list:
        losshistory, train_state = model.train(iterations=iterations)

        dde.saveplot(losshistory, train_state, issave=True, isplot=True)
        test_loss = train_state.best_loss_test

        end_time = time.time()
        elapsed_time = end_time - start_time
        print("Elapsed time:", elapsed_time, "seconds")

        N = 256
        x0 = np.linspace(0,L,N)

        X = x0.reshape(-1,1)

        Ys = model.predict(X)


        Y10 = Ys[:,0]
        Y20 = Ys[:,1]
        Y30 = Ys[:,2]
        Y40 = Ys[:,3]

        N1 = round(N * (pl1 / L))
        N2 = round(N * (pl2 / L))
        N3 = round(N * (pl3 / L))

        Y1 = Y10[:N1]
        Y2 = Y20[N1:N2]
        Y3 = Y30[N2:N3]
        Y4 = Y40[N3:]


        Y = np.hstack([Y1,Y2,Y3,Y4])

        plt.plot(x0, Y)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Plot of y = ODE solution')
        plt.show()

        j = [func(x0i) for x0i in x0]

        indices = np.linspace(0, len(x0)-1, 10, dtype=int)  # Get the indices for 10 equally spaced numbers
        #selected_numbers = k[indices]  # Select the numbers at the corresponding indices

        plt.plot(x0, j)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Function plot')
        plt.show()

        selected_elements_pinn = [Y[i] for i in indices]
        selected_elements_ana = [j[i] for i in indices]

        differences = [abs(p - a) for p, a in zip(selected_elements_pinn, selected_elements_ana)]  # Calculate differences
        sum_of_differences = sum(differences)/abs(sum(selected_elements_ana))  # Sum all differences

        print("Differences:", differences)
        print("Sum of differences:", sum_of_differences)


        # Store the results in a dictionary
        result = {
            'iterations': iterations,
            'elapsed_time': elapsed_time,
            'sum_of_differences': sum_of_differences,
            'test_loss': test_loss
        }
        results.append(result)
        losses.append(test_loss)

        print(f"Iterations: {iterations}, Elapsed time: {elapsed_time:.2f} seconds")
        print(f"Sum of differences: {sum_of_differences:.2e}")
        print(f"Test loss: {test_loss:.2e}")
        print("")


# Convert results and losses to pandas DataFrames
results_df = pd.DataFrame(results)

# Save the results to an Excel file

with pd.ExcelWriter('/content/results.xlsx') as writer:
    results_df.to_excel(writer, sheet_name='Results_oneload', index=False)
```
